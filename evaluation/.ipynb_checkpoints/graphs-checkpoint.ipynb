```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import os
```

## Generate Graphs

```python
benchmarks = ['kafka', 'tomcat', 'h2', 'lusearch']
gcs = ['serial', 'parallel', 'g1', 'go']
gcs_labels = ['Serial', 'Parallel', 'G1', 'GoGC']
dump_dirs = ['memory_pressure', 'no_memory_pressure']
```

```python
for dump_dir in dump_dirs:
    for benchmark in benchmarks:
        # Create a figure and axis object
        fig, ax = plt.subplots(figsize=(12, 6))
        line_styles = ['-', '--', ':', '-.', 'dotted']
        markers = ['x', 'o', '^', 's', 'd'] 
        memory_pressure_file_path = f'{dump_dir}/{benchmark}'
        for i, gc in enumerate(gcs):
            file = f'scratch_{gc}/dacapo-latency-usec-simple-0.csv'
            path = os.path.join(memory_pressure_file_path, file)
            df = pd.read_csv(path, header=None)
            df.columns = ["start_time", "end_time"]
            df['latency'] = df['end_time'] - df['start_time']
            df['latency'] = df['latency']/1000
            ax.plot(df['latency'][20000:25000], label=gcs_labels[i], linestyle=line_styles[i], linewidth=1, marker=markers[i], markersize=4)
    
        # Set the x-axis label
        ax.set_xlabel('Request Number')
        
        # Set the y-axis label
        ax.set_ylabel('Latency')
        
        # Set the title
        # ax.set_title(f'{benchmark} benchmark')
        
        # Add a legend
        ax.legend()
    
        plt.savefig(f'graphs/{dump_dir}/{benchmark}_latency.png')
        # Show the plot
        plt.show()
```

## Filter Logs

```python
benchmarks = ['kafka', 'tomcat', 'h2', 'lusearch', 'graphchi']
gcs = ['serial', 'parallel', 'g1', 'go', 'epsilon']
dump_dirs = ['memory_pressure', 'no_memory_pressure']
```

```python
import re

def parse_log_file(file_path, benchmark):
    total_pause_time = 0
    start_parsing = False
    completion_time = 0
    count = 0
    with open(file_path, 'r') as file:
        for line in file:
            if f'=====' in line:
                start_parsing = True
            if start_parsing and'Pause' in line:
                pause_time = extract_pause_time(line)
                total_pause_time += pause_time
                count += 1
            if 'PASSED' in line:
                start_parsing = False
                pattern = r'PASSED in (\d+) msec'
                match = re.search(pattern, line)
                if match:
                    completion_time = int(match.group(1))
    average_pause_time = total_pause_time/count
    return total_pause_time, average_pause_time, completion_time

def extract_pause_time(line):
    pattern = r'(\d+\.\d+)ms'
    match = re.search(pattern, line)
    if match:
        return float(match.group(1))
    else:
        return 0

def extract_completion_time(line):
    pattern = r'(\d+) msec'
    match = re.search(pattern, line)
    if match:
        return float(match.group(1))
    else:
        return 0

def parse_epsilon_log(file_path):
    total_time = 0
    count = 0
    start_parsing = False
    with open(file_path, 'r') as file:
        for line in file:
            if f'=====' in line:
                start_parsing = True
            if start_parsing and 'Lisp2-style Mark-Compact' in line:
                pause_time = extract_pause_time(line)
                total_time += pause_time
                count += 1
            if 'PASSED' in line:
                start_parsing = False
                pattern = r'PASSED in (\d+) msec'
                match = re.search(pattern, line)
                if match:
                    completion_time = int(match.group(1))

    return total_time, total_time/count, completion_time
```

```python
for dump_dir in dump_dirs:
    print('-'*10, dump_dir, '-'*10)
    for benchmark in benchmarks:
        print('='*5, benchmark, '='*5)
        memory_pressure_file_path = f'{dump_dir}/{benchmark}'
        for i, gc in enumerate(gcs):
            print('_'*3, gc, '_'*3)
            file = f'dump_{gc}.txt'
            path = os.path.join(memory_pressure_file_path, file)
            if gc == 'epsilon':
                total_pause_time, avg_pause_time, completion = parse_epsilon_log(path)
            else:
                total_pause_time, avg_pause_time, completion = parse_log_file(path, benchmark)
            print(f"Total time spent in GC: {total_pause_time:.2f} ms")    
            print(f"Average time spent in GC: {avg_pause_time:.2f} ms")
            print(f"Completion time: {completion} ms")
            print(f"% time spent in GC: {((total_pause_time/completion)*100):.2f}")
```
